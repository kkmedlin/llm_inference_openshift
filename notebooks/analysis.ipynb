```python
import pandas as pd
import matplotlib.pyplot as plt

# Load benchmark results
df = pd.read_csv("../benchmarks/results.csv")

# Make sure latency is numeric
df['latency_sec'] = pd.to_numeric(df['latency_sec'], errors='coerce')

# --- Latency Analysis ---
plt.figure(figsize=(8,5))
df.groupby("prompt")["latency_sec"].mean().plot(
    kind="bar", color="skyblue"
)
plt.ylabel("Latency (seconds)")
plt.xlabel("Prompt")
plt.title("Average CPU Inference Latency per Prompt")
plt.show()

# --- Review Generated Outputs ---
print("\nGenerated Outputs:\n")
for idx, row in df.iterrows():
    print(f"Prompt: {row['prompt']}")
    print(f"Generated text: {row['output_text']}\n{'-'*50}\n")

```
